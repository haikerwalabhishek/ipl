```python
#columns to remove
#mid,venue does not matter in winning
col=['mid','venue','batsman','bowler','striker','non-striker']
data.drop(columns=col,axis=1,inplace=True)
```

```python
data.head()
```

```python
data['bat_team'].unique()
```

```python
data['bat_team']=data['bat_team'].replace('Rising Pune Supergiant','Rising Pune Supergiants')
```

```python
data['bowl_team'].unique()
```

```python
consistent_team=['Royal Challengers Bangalore', 'Kings XI Punjab',
       'Delhi Daredevils', 'Kolkata Knight Riders', 'Rajasthan Royals',
       'Mumbai Indians', 'Chennai Super Kings', 'Deccan Chargers',
       'Pune Warriors', 'Kochi Tuskers Kerala', 'Sunrisers Hyderabad']
```

```python
data=data[(data['bat_team'].isin(consistent_team)) &(data['bowl_team'].isin(consistent_team))]
```

```python
data.shape
```

```python
#removing first five overs from the data
data=data[data['overs']>5.0]
```

```python
data.head()
```

```python
#converting datetime 
data['date']=pd.to_datetime(data['date'])
```

```python
#categorical columns
encoded_df=pd.get_dummies(data=data,columns=['bat_team','bowl_team'])
```

```python
# encoded_df = encoded_df[['date', 'bat_team_Chennai Super Kings', 'bat_team_Delhi Daredevils', 'bat_team_Kings XI Punjab',
#               'bat_team_Kolkata Knight Riders', 'bat_team_Mumbai Indians', 'bat_team_Rajasthan Royals',
#               'bat_team_Royal Challengers Bangalore', 'bat_team_Sunrisers Hyderabad',
#               'bowl_team_Chennai Super Kings', 'bowl_team_Delhi Daredevils', 'bowl_team_Kings XI Punjab',
#               'bowl_team_Kolkata Knight Riders', 'bowl_team_Mumbai Indians', 'bowl_team_Rajasthan Royals',
#               'bowl_team_Royal Challengers Bangalore', 'bowl_team_Sunrisers Hyderabad',
#               'overs', 'runs', 'wickets', 'runs_last_5', 'wickets_last_5', 'total']]
```

```python
pd.set_option('display.max_columns',100)
encoded_df.head()
```

```python
x_train=encoded_df.drop(labels='total',axis=1)[encoded_df['date'].dt.year<=2016]
x_test=encoded_df.drop(labels='total',axis=1)[encoded_df['date'].dt.year>2016]
```

```python
y_train=encoded_df[encoded_df['date'].dt.year<=2016]['total']
y_test=encoded_df[encoded_df['date'].dt.year>2016]['total']
```

```python
y_train1=encoded_df[encoded_df['date'].dt.year<=2016]['total'].values
y_test1=encoded_df[encoded_df['date'].dt.year>2016]['total'].values
```

```python
#removing date column
x_train.drop('date',axis=1,inplace=True)
x_test.drop('date',axis=1,inplace=True)
```

```python
#model building
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,reg.predict(x_test))
```

```python
import pickle
pickle.dump(reg,open('ipl_reg.pkl','wb'))
```
